```python
def sigmoid ( z ) :
    return 1 / (1 + np . exp ( - z ) )

def sigmoid_derivative ( z ) :
    s = sigmoid ( z )
    return s * (1 - s )
def mse_derivative ( y_hat , y ) :
    return 2 * ( y_hat - y )

x = np . array ([1 , 0])
y = 1
w = np . array ([ -1 , 2])

# Forward pass
z = np . dot (w , x )


y_hat = sigmoid ( z )
loss = ( y_hat - y ) ** 2
print (" ===␣ Forward ␣ Pass ␣=== ")
print ( f"z␣=␣{z}")
print ( f" y_hat ␣=␣{ y_hat }")
print ( f" loss ␣=␣{ loss }")

# Backward pass
dloss_dyhat = mse_derivative ( y_hat , y )
dyhat_dz = sigmoid_derivative ( z )
dloss_dz = dloss_dyhat * dyhat_dz
dloss_dw1 = dloss_dz * x [0]
dloss_dw2 = dloss_dz * x [1]

print ("\n ===␣ Backward ␣ Pass ␣=== ")
print ( f" dloss / dy_hat ␣=␣{ dloss_dyhat }")
print ( f" dy_hat /dz␣=␣{ dyhat_dz }")
print ( f" dloss /dz␣=␣{ dloss_dz }")
print ( f" dloss / dw1␣=␣{ dloss_dw1 }")
print ( f" dloss / dw2␣=␣{ dloss_dw2 }")
```
